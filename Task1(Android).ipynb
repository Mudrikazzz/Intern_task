```python
df["Price"]=df["Price"].astype(str).str.replace("$","")
```

```python
df['Installs']=df['Installs'].astype(str).str.replace("+","")
```

```python
df['Installs']=df['Installs'].astype(str).str.replace(",","")
```

```python
df["Price"]=df["Price"].astype("float")
df["Installs"]=df["Installs"].astype("float")
```

```python
no_of_cat=df["Category"].value_counts().reset_index()
no_of_cat.head()
```

```python
plt.figure(figsize=(18,8))
sns.barplot(no_of_cat,x="Category",y="count",palette='Greens')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(df["Rating"],binwidth=0.1)
plt.axvline(x=(df["Rating"].mean()),linestyle='--',label=f'Mean:{df["Rating"].mean():.2f}',color='red')
plt.legend()
plt.show()
```

```python
no_type=df["Type"].value_counts().values
labels=df["Type"].unique()
```

```python
plt.pie(no_type,labels=labels,autopct='%1.3f%%')
plt.show()
```

```python
pri_rate=df.groupby("Price")["Rating"].mean().reset_index()
sns.regplot(x='Price',y='Rating',data=pri_rate,scatter_kws={"s":30},line_kws={"color":"red"})
plt.show()
```

```python
rate_cat=df.groupby("Category")["Rating"].mean().reset_index()
sns.barplot(x="Category",y="Rating",data=rate_cat,color='red')
plt.xticks(rotation=90)
plt.show()
```
